```python
from brian2 import *

# Configuración de 2 Entradas y 1 Neurona de Salida (Sistema jerárquico)
tau = 10*ms
eqs = 'dv/dt = (-70*mV - v + I) / tau : volt'

# Dos redes de entrada independientes
Entrada_A = NeuronGroup(1, eqs, threshold='v > -50*mV', reset='v = -80*mV', method='exact')
Entrada_B = NeuronGroup(1, eqs, threshold='v > -50*mV', reset='v = -80*mV', method='exact')

# La neurona de salida (El "Detector")
Salida = NeuronGroup(1, eqs, threshold='v > -50*mV', reset='v = -80*mV', method='exact')

# Acoplamiento: Cada entrada por sí sola NO es suficiente para disparar la salida
# (Necesitan sumar sus voltajes -> Lógica de Red)
S = Synapses(NeuronGroup.union(Entrada_A, Entrada_B), Salida, on_pre='v_post += 12*mV')
S.connect()

# Monitores
m = SpikeMonitor(Salida)
state = StateMonitor(Salida, 'v', record=0)

# Simulamos pulsos en tiempos distintos y luego juntos
Entrada_A.v = -50.1*mV # Forzamos un disparo de A
run(20*ms)
Entrada_B.v = -50.1*mV # Forzamos un disparo de B
run(20*ms)
# Disparo simultáneo (Coincidencia)
Entrada_A.v = -50.1*mV; Entrada_B.v = -50.1*mV
run(20*ms)

plot(state.t/ms, state.v[0]/mV)
plt.axhline(-50, color='r', ls='--', label='Umbral')
plt.title("Detección de Coincidencia: La unidad básica de una Red")
plt.legend()
plt.show()
```
