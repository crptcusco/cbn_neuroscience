```python
from brian2 import *
import matplotlib.pyplot as plt

start_scope()

# Modelo LIF mínimo
eqs = '''
dv/dt = (-(v + 65*mV) + 10*mV) / (20*ms) : volt
'''

G = NeuronGroup(
    1,
    eqs,
    threshold='v > -50*mV',
    reset='v = -65*mV',
    method='euler'
)

G.v = -65*mV

M = StateMonitor(G, 'v', record=True)
S = SpikeMonitor(G)

run(100*ms)

plt.plot(M.t/ms, M.v[0]/mV)
plt.axhline(-50, linestyle='--')
plt.xlabel("ms")
plt.ylabel("mV")
plt.title("Paso 2 – Neurona LIF (ultra simple)")
plt.show()

print("Spikes:", S.count[0])
```
